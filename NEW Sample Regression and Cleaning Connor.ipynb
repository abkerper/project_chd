##Cleaning and Sample Regression

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv("/content/fhs_train.csv")
df.head()
```

```python
df.describe()
```

```python
variables = ['age','sysBP','totChol','BMI','TenYearCHD','BPMeds']
train_df = df.loc[:,variables]
train_df.head()
```

```python
train_df.isnull().sum(axis = 0)
```

```python
clean_df = train_df.dropna()
clean_df.describe()
```

Has consistent row counts meaning all NAN values have been dropped

```python
y = clean_df['TenYearCHD']
X = clean_df.loc[:,['age','sysBP','totChol','BMI']]
reg = LinearRegression().fit(X, y) # Fit the linear model
print('Intercept Coefficent:', reg.intercept_) # Intercept value
print('Coefficents:' , reg.coef_) # Regression coefficients
print('R Squared:',reg.score(X, y))
```

```python
y = clean_df['TenYearCHD']
X = clean_df.loc[:,['age','sysBP','totChol','BMI','BPMeds']]
reg = LinearRegression().fit(X, y) # Fit the linear model
print('Intercept Coefficent:', reg.intercept_) # Intercept value
print('Coefficents:' , reg.coef_) # Regression coefficients
print('R Squared:',reg.score(X, y))
```

```python
train_df = pd.read_csv('/content/fhs_train.csv')


test_df = pd.read_csv('/content/fhs_test.csv')

train_df = train_df.dropna()
test_df = test_df.dropna()


X_train = train_df[['age', 'sysBP', 'totChol', 'BMI']]
y_train = train_df['TenYearCHD']


X_test = test_df[['age', 'sysBP', 'totChol', 'BMI']]
y_test = test_df['TenYearCHD']
```

```python
reg = LinearRegression()

# Step 5: Fit the model to the training data
reg.fit(X_train, y_train)

# Step 6: Make predictions on the testing data
y_pred = reg.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse:.4f}')
print(f'R^2 Score: {r2:.4f}')
```
